```python
%matplotlib inline
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class LocalLevel :
    
    def __init__(self,epsilon,snr):
        
        self.Z      = np.array([1])
        self.T      = np.array([1])
        self.R      = np.array([1])
        self.Tshock = lambda : np.random.normal(0,np.sqrt(epsilon))
        self.Pshock = lambda : np.random.normal(0,np.sqrt(snr*epsilon))
        self.H      = np.array([epsilon])
        self.Q      = np.array([snr*epsilon])
        
    def emulate(self,interactions):
        
        alpha = np.random.normal(0,10)
        y = np.zeros(interactions)
        for i in range(interactions):
            alpha = self.T*alpha + self.R*self.Pshock()
            y[i]  = self.Z*alpha + self.Pshock()
        return y
    
    def filter(self,series):
        
        a     = 0
        P     = 10**7
        F     = 0
        level = np.zeros(len(series))
        
        for i in range(len(series)):
            
            F    = self.Z*P*self.Z.t() + self.H
            inov = series[i] - self.Z*a
            a    = self.T*a + self.T*self.Z*P*F*inov
           # P    = 
            
        return None 
        
```

# Primeira Parte : Simulação

```python
def PlotSimulations(epsilon,snr,nr_of_simulations):
    ll = LocalLevel(epsilon,snr)
    series_dict = { 'simulation ' + str(i) : ll.emulate(300)   for i in range(nr_of_simulations) }
    df = pd.DataFrame(series_dict)
    df.plot(title = 'Epsilon = ' + str(epsilon) + ' , Snr = ' + str(snr))
```

```python
PlotSimulations(1,10,5)
```
